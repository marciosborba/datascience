# Análise de dados através de API

link ->https://dadosabertos.camara.leg.br/api/v2/partidos?itens=100&ordem=ASC&ordenarPor=sigla


```python
# Importar as bibliotecas

import pandas as pd

# Bibliteca requests requisita um endereço http online

import requests
```

```python
# Chamar o matplotlib inline

%matplotlib inline
```

```python
# Obter os dados da API

# Criar uma variavel para URL

url = 'https://dadosabertos.camara.leg.br/api/v2/partidos?itens=100&ordem=ASC&ordenarPor=sigla'
```

```python
# Obtendo os dados

resposta = requests.get(url)
```

```python
# Verificando a resposta

resposta.text
```

```python
# Converter resposta do API em DataFrame

df= pd.DataFrame(resposta.json()['dados'])
```

```python
df.head(50)
```

```python
# Mostrar numero de partidos no arquivo

df['sigla'].count()
```

```python
# Nome

df['nome']
```

```python
# Qualquer campo do DataFrame pode ser obtido com a função de localizar ( loc e Iloc)

# O iloc retorna um campo com base no seu indice (linha, coluna)

df.iloc[38, 1]
```

```python
# O loc retorna o campo com base no nome da coluna

df.loc[5, 'nome']
```
