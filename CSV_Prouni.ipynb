```python
# Carregar Arquivo CSV

df = pd.read_csv('cursosProuni.csv')

df.head()
```

```python
# Análise matemática das colunas numéricas

df.describe()
```

```python
# Descrever dados

df.info()
```

```python
# Quantidade de curso por UF

df['uf_busca'].value_counts()
```

```python
# Qual o curso com a mensalidade mais alta

df[df['mensalidade']== df['mensalidade'].max()]
```

```python
# Qual a mensalidade mais barata do Brasil

df[df['mensalidade']== df['mensalidade'].min()]
```

```python
# Mostrar mensalidade mais cara por UF

df.groupby('uf_busca')['mensalidade'].max()
```

```python
# Mostrar mensalidade mais barata por UF

df.groupby('uf_busca')['mensalidade'].min()
```

```python
# Mostrar todos os dados dos cursos mais caros por UF

df.loc[df.groupby('uf_busca')['mensalidade'].idxmax()]
```

```python
# Mostrar mensalidade entre uma faixa de valor ( 500 a 1000 reais)

df[(df['mensalidade'] >500) & df['mensalidade'] < 1000]
```

```python
# Quais as 5 maiores mensalidades

df.nlargest(5, 'mensalidade')
```

```python
# Mostrar as 5 menores mensalidades

df.nsmallest(6,'mensalidade')
```
