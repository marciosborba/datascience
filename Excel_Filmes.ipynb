# Análise de filmes exibidos no Brasil de 2009 a 2018

Com base nos dados da Ancine (https://oca.ancine.gov.br/cinema)

Arquivo -> 2120.xlsx

```python
# Importar as bibliotecas

import pandas as pd

# Configurar o Matplotlib

%matplotlib inline
```

```python
# Criar o DataFrame com base no arquivo excel
# Descartar as duas primeiras linhas do arquivo (skiprows)

df= pd.read_excel('2120.xlsx', skiprows=2, skipfooter=16)

# Mostrar as linhas iniciais do arquivo

df.head()

```

```python
# Mostrar o final do arquivo

df.tail(2)
```

```python
# Verificar todos os tipos de colunas

df.dtypes
```

```python
# Converter o campo STR em INT
df['Data de lançamento'] = pd.to_datetime(df['Data de lançamento'])
```

```python
# Encontrar o erro da coluna

df['Data de lançamento'].value_counts()
```

```python
# Exibir alguns filmes que estão com data == relançamento

df[df['Data de lançamento'] == 'Relançamento'].head()
```

# Decisão sobre dados incorretos/faltantes/divergentes

```python
## Verificar tamanho do DataFrame

df.shape
```

```python
# Criar um novo DataFrame sem os relançamentos
df_novo = df[df["Data de lançamento"]!= 'Relançamento'].copy()
```

```python
# Verificar o novo DataFrame

df_novo.dtypes
```

```python
# Converter o campo STR em INT

df_novo['Data de lançamento'] = pd.to_datetime(df_novo['Data de lançamento'])
```

```python
df_novo.dtypes
```

```python
# Quais os anos tiveram mais filmes lançados

df_novo['Data de lançamento'].dt.year.value_counts()
```

```python
# Plotar um grafico de barras

df_novo['Data de lançamento'].dt.year.value_counts().plot.bar()
```

```python
# Qual o filme teve a maior bilheteria

df_novo[df_novo['Renda (R$) no ano de exibição']== df_novo['Renda (R$) no ano de exibição'].max()]
```

```python
# Os tres filmes com maior bilheteria

df_novo.nlargest(3, 'Renda (R$) no ano de exibição')
```

```python
# Qual o filme com menor bilheteria

df_novo[df_novo['Renda (R$) no ano de exibição']== df_novo['Renda (R$) no ano de exibição'].min()]
```

```python
# Quantos filmes nacionais e estrangeiros
 
df_novo['Nacionalidade da obra'].value_counts()    
```
